```python
# Rule-Based Chatbot: College Library Assistant

def library_chatbot():
    print(" Welcome to College Library Assistant Chatbot!")
    print("Type 'exit' to end the chat.\n")

    while True:
        try:
            user_input = input("You: ").lower()

            # Exit condition
            if user_input == "exit":
                print("Bot: Thank you! Visit again ")
                break

            # Greeting
            elif "hello" in user_input or "hi" in user_input:
                print("Bot: Hello! How can I assist you today?")

            # Library timing
            elif "timing" in user_input:
                print("Bot: Library is open from 9 AM to 8 PM.")

            # Book availability
            elif "book" in user_input:
                print("Bot: Please tell me the book name.")
                book_name = input("You: ")
                print(f"Bot: Checking availability for '{book_name}'...")
                print("Bot: Yes, the book is available.")

            # Fine calculation (Mathematical)
            elif "fine" in user_input:
                print("Bot: Enter number of late days:")

                days = int(input("You: "))

                fine_per_day = 5
                total_fine = days * fine_per_day

                print(f"Bot: Your total fine is ₹{total_fine}")

            # Issue book condition
            elif "issue" in user_input:
                print("Bot: Enter number of books you already have:")

                books = int(input("You: "))

                if books < 3:
                    print("Bot: You can issue more books.")
                else:
                    print("Bot: You have reached the maximum limit (3 books).")

            # Return book
            elif "return" in user_input:
                print("Bot: Book returned successfully. Thank you!")

            # Default response
            else:
                print("Bot: Sorry, I didn't understand that. Please try again.")

        except ValueError:
            print("Bot: Please enter valid numbers only.")

        except Exception as e:
            print("Bot: Something went wrong:", e)

# Run chatbot
library_chatbot()
```
